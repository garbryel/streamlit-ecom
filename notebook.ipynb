### 4. Seberapa sering seorang pelanggan melakukan pembelian dalam beberapa bulan terakhir?

#### Menggunakan RFM Analysis

- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
- Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
- Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
all_order_df['order_purchase_timestamp'] = pd.to_datetime(all_order_df['order_purchase_timestamp'])

# Hitung Recency
current_date = all_order_df['order_purchase_timestamp'].max()
recency_df = all_order_df.groupby('customer_id')['order_purchase_timestamp'].max().reset_index()
recency_df['Recency'] = (current_date - recency_df['order_purchase_timestamp']).dt.days

# Hitung Frequency
frequency_df = all_order_df.groupby('customer_id').size().reset_index(name='Frequency')

# Hitung Monetary
monetary_df = all_order_df.groupby('customer_id')['price'].sum().reset_index()

# Gabungkan ketiganya
rfm_df = pd.merge(recency_df[['customer_id', 'Recency']], frequency_df, on='customer_id')
rfm_df = pd.merge(rfm_df, monetary_df.rename(columns={'price': 'Monetary'}), on='customer_id')

print(rfm_df.head())
```

```python
# Visualize Recency
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.hist(rfm_df['Recency'], bins=20, color='skyblue', edgecolor='black')
plt.title('By Recency (days)')

# Visualize Frequency
plt.subplot(1, 3, 2)
plt.hist(rfm_df['Frequency'], bins=20, color='skyblue', edgecolor='black')
plt.title('By Frequency')

# Visualize Monetary
plt.subplot(1, 3, 3)
plt.hist(rfm_df['Monetary'], bins=20, color='skyblue', edgecolor='black')
plt.title('By Monetary')

#plt.tight_layout()
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```

Dari visualisasi di atas dapat dilihat frekuensi pembelian dominan 1, hal ini dapat disimpulkan dalam sebulan costumer tidak terlalu sering untuk melakukan pembelian berulang.
